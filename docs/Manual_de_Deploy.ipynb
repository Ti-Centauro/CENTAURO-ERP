# Passo a Passo para Configurar
**(Faça isso após terminarmos a refatoração do Backend)**

Como você tem uma pasta para frontend e outra para backend no mesmo projeto (Monorepo), a configuração tem um "pulo do gato":

## 1. Preparar o Repositório (GitHub)
Garanta que todo o seu código (backend e frontend) está no GitHub.

> **Dica:** Crie um arquivo `vercel.json` na raiz ou na pasta `frontend`, e um `Procfile` ou `Dockerfile` na pasta `backend`. (Posso gerar esses arquivos para você depois).

## 2. Configurar o Railway (Backend + Banco)
O Railway é inteligente. Ele vai ler sua pasta backend.

1. Crie conta no Railway (use o GitHub Login).
2. Clique em "New Project" -> "Deploy from GitHub repo".
3. Selecione o repositório do `CENTAURO-ERP-2`.

### O Pulo do Gato
Nas configurações do serviço, defina o **"Root Directory" (Diretório Raiz)** como `/backend`. Assim ele ignora a pasta do React.

### Banco de Dados
No Railway, adicione um serviço "PostgreSQL".

> **Atenção:** Seu código hoje usa SQLite. Vamos precisar mudar uma linha no arquivo `.env` para ele conectar no Postgres do Railway. É super simples.

## 3. Configurar a Vercel (Frontend)
A Vercel é nativa para React/Vite.

1. Crie conta na Vercel (use o GitHub Login).
2. Clique em "Add New..." -> "Project".
3. Importe o repositório `CENTAURO-ERP-2`.

### O Pulo do Gato
Na tela de importação, clique em "Edit" na opção **"Root Directory"** e selecione a pasta `frontend`.

A Vercel vai detectar que é Vite e configurar o comando de build (`npm run build`) sozinha.

## 4. A Conexão (Variáveis de Ambiente)
Para eles conversarem, eles precisam saber o endereço um do outro.

### Na Vercel (Settings > Environment Variables):
Crie uma variável `VITE_API_URL` e coloque o link que o Railway gerou (ex: `https://centauro-backend-production.railway.app`).

### No Railway (Variables):
O Railway já injeta as variáveis do Banco de Dados (Postgres) automaticamente. Só precisaremos ajustar seu `database.py` para ler isso.

## Respondendo sua preocupação: "É complicado fazer alterações rápidas?"
Não, é a coisa mais rápida do mundo.

**Cenário:** Você achou um bug no texto de um botão.

1. Abre a IDE.
2. Corrige o texto.
3. Clica no ícone do Git na IDE -> Digita "fix: texto botão" -> Clica em "Sync Changes" (ou Push).

**Fim.** Você pode ir tomar um café. Em cerca de 2 minutos, o site oficial estará atualizado.

> **Vantagem extra:** Se você errar algo e o site quebrar, tanto a Vercel quanto o Railway têm um botão "Rollback" (Voltar Versão) que reverte tudo em 10 segundos.

## O que precisamos fazer no código ANTES do deploy?
Para isso funcionar liso, precisamos de 3 ajustes técnicos que faremos via prompt com a IA depois:

1. **Trocar SQLite por Postgres:** O SQLite é um arquivo local. Na nuvem (Railway), cada vez que você fizer deploy, o arquivo seria apagado se não configurarmos volumes (o que é chato). Usar o Postgres do Railway é a solução profissional.
2. **CORS:** Precisamos avisar o Backend (FastAPI) para aceitar requisições do site da Vercel (hoje ele deve estar aceitando só localhost).
3. **Requirements.txt:** Garantir que todas as bibliotecas (pandas, openpyxl, sqlalchemy) estejam listadas para o Railway instalar.
